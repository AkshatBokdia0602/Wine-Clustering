# Preprocessing the Data

```python
X.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 10)
X_test = pca.fit_transform(X)
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth = 2, color = 'blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components = 3)
X_pca = pca.fit_transform(X)
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', n_init = "auto")
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss, 'o-', linewidth = 2, color = 'red')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', n_init = "auto")
y_kmeans = kmeans.fit_predict(X_pca)
```

```python
unique, counts = np.unique(y_kmeans, return_counts = True)
print(np.asarray((unique, counts)).T)
```

```python
plt.figure(figsize=(6,6))
ax = plt.axes(projection = '3d')
ax.scatter(X_pca[y_kmeans == 0, 0], X_pca[y_kmeans == 0, 1], X_pca[y_kmeans == 0, 2], s = 100, c = '#16A37A', label = 'Cluster 1', alpha = 0.5, marker='^')
ax.scatter(X_pca[y_kmeans == 1, 0], X_pca[y_kmeans == 1, 1], X_pca[y_kmeans == 1, 2], s = 100, c = '#B319FA', label = 'Cluster 2', alpha = 0.5, marker='+')
ax.scatter(X_pca[y_kmeans == 2, 0], X_pca[y_kmeans == 2, 1], X_pca[y_kmeans == 2, 2], s = 100, c = '#FAB832', label = 'Cluster 3', alpha = 0.5, marker='*')
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s = 250, c = 'black', label = 'Centroids')

ax.set_xlabel('PC1', rotation=-30)
ax.set_ylabel('PC2', rotation=45)
ax.set_zlabel('PC3', rotation=90)
ax.zaxis.labelpad = -4
plt.legend(bbox_to_anchor = (1.0, 1.0), ncol = 2)
plt.title("Clusters of Wine (PCA)")
plt.show()
```

## Agglomerative Clustering

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X_pca, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Wine')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X_pca)
```

```python
unique, counts = np.unique(y_hc, return_counts = True)
print(np.asarray((unique, counts)).T)
```

```python
plt.figure(figsize=(6,6))
ax = plt.axes(projection = '3d')
ax.scatter(X_pca[y_hc == 0, 0], X_pca[y_hc == 0, 1], X_pca[y_hc == 0, 2], s = 100, c = '#F0372B', label = 'Cluster 1', alpha = 0.5)
ax.scatter(X_pca[y_hc == 1, 0], X_pca[y_hc == 1, 1], X_pca[y_hc == 1, 2], s = 100, c = '#EFB600', label = 'Cluster 2', alpha = 0.5)
ax.scatter(X_pca[y_hc == 2, 0], X_pca[y_hc == 2, 1], X_pca[y_hc == 2, 2], s = 100, c = '#1390F0', label = 'Cluster 3', alpha = 0.5)

ax.set_xlabel('PC1', rotation = -30)
ax.set_ylabel('PC2', rotation = 45)
ax.set_zlabel('PC3', rotation = 90)
ax.zaxis.labelpad = -4
plt.legend(bbox_to_anchor = (1.0, 1.0), ncol = 2)
plt.title("Clusters of Wine (PCA)")
plt.show()
```
